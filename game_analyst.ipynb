```python
import pandas as pd
df = pd.read_csv('vgsales.csv')

# Data Cleaning
df['Year'] = df['Year'].fillna(df['Year'].median())
df['Publisher']= df['Publisher'].fillna('Unknown')
```

```python
df.isna().sum()
```

```python
# Pengelompokan Penjualan Berdasarkan penjualan Global
df_sales_kategory =  pd.cut (
    df['Global_Sales'],
    bins=[0,30,50, df['Global_Sales'].max()],
    labels=['Low', 'Medium', 'High'],
    include_lowest=True
)

df['Sales_Category'] = df_sales_kategory

# Kateghori Tahun Rilis 
df_era_category = pd.cut(
    df['Year'],
    bins= [1965, 2005, 2018, 2026],
    labels=['Old', 'Modern', 'Recent']
)

df['Year_Category'] = df_era_category

# Main Region 
df['Main_Region'] = df[
    ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']
].idxmax(axis=1)

df['Main_Region'] = df['Main_Region'].str.replace('_Sales', '')
# Kategory Publisher
publisher_sales = df.groupby('Publisher')['Global_Sales'].transform('sum')

df['Publisher_Size'] = pd.cut(
    publisher_sales,
    bins=[0, 50, 200, publisher_sales.max()],
    labels=['Small', 'Medium', 'Big']
)
```
